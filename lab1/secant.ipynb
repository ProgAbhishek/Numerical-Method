```python
def f(x):
    return x**3 - 4*x - 9

def secant(x0, x1, tol=0.0005, max_iter=100):
    for i in range(max_iter):
        f0 = f(x0)
        f1 = f(x1)

        if abs(f1 - f0) < 1e-10:
            print("Division by near-zero â€” stopping.")
            return None

        x2 = x1 - f1 * (x1 - x0) / (f1 - f0)

        if abs(x2 - x1) < tol:
            print(f"The correct root up to 3 decimal places is {x2:.4f}")
            return x2

        x0, x1 = x1, x2

    print("Method did not converge.")
    return None

# Input
x0 = float(input("Enter the first initial value (x0): "))
x1 = float(input("Enter the second initial value (x1): "))

secant(x0, x1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_val = np.linspace(x0- 1, x1 + 1, 400)
y_val = f(x_val)

plt.figure(figsize=(10, 6))
plt.plot(x_val, y_val, label="f(x)", color='blue')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Secant Method Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
