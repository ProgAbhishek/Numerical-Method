```python
def lagrange(x,y,xp):
    n=len(x)
    yp=0

    for i in range(n):
        p=1
        for j in range(n):
            if i != j:
                p *= (xp-x[j])/(x[i]-x[j])
            yp += p*y[i]
        return yp

#data
x = [5,10,15,20]
y = [12,15,20,27]

xp = float(input("Enter x to interpolate at : "))
yp = lagrange(x,y,xp)

print(f"Interpolated value at x= {xp} is {yp:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_vals = np.linspace(min(x), max(x), 200)
y_vals = [lagrange(x, y, xi) for xi in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="Lagrange Interpolation", color='blue')
plt.scatter(x, y, color='red', label="Data Points", zorder=5)
plt.scatter(xp, yp, color='green', label=f"Interpolated Point\n({xp:.2f}, {yp:.2f})", zorder=6)
plt.title("Lagrange Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
