```python
def f(x):
    return x**3-4*x-9

def bisection(a,b,tol=0.0005):
    fa=f(a)
    fb=f(b)

    if fa * fb > 0 :
        print("The soln doesnt exist")
        return None

    while True:
        c=(a+b)/2
        fc=f(c)

        if abs(fc)<tol:
            break
    
        if fc*fa>0:
            a=c
            fa=fc
        else:
            b=c
            fb=fc

    print(f"The correct root upto 3 decimal placce is {c:.4f}")
    return c

#input
a= float(input("Enter the initial value of a : "))
b= float(input("Enter the initial value of b : "))

bisection(a,b)
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_val=np.linspace(a-1,b+1,400)
y_val=f(x_val)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_val,y_val,label="f(x)")
plt.axhline(0,color='gray',linestyle='--')
plt.title('Bisection method Visualization')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
