```python
import math
def exponential_fit(x, y):
    n = len(x)
    ln_y = [math.log(yi) for yi in y]

    sumx = sum(x)
    sumlny = sum(ln_y)
    sumx2 = sum(xi * xi for xi in x)
    sumxlny = sum(x[i] * ln_y[i] for i in range(n))

    d = (n * sumx2) - (sumx ** 2)
    d1 = (sumlny * sumx2) - (sumx * sumxlny)
    d2 = (n * sumxlny) - (sumx * sumlny)

    if d == 0:
        print("Error: Division by zero in calculation.")
        return None

    A = d1 / d
    b = d2 / d
    a = math.exp(A)

    print(f"The best fit exponential curve is: y = {a:.3f} * e^({b:.3f}x)")
    return a, b

# Sample data
x = [1, 2, 3, 4]
y = [2.7, 7.4, 20.1, 54.6]

# Fit the exponential curve
a, b = exponential_fit(x, y)
```

```python
import matplotlib.pyplot as plt
x_vals = [i for i in range(min(x), max(x) + 1)]
y_vals = [a * math.exp(b * xi) for xi in x_vals]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_vals, y_vals, color='red', label='Exponential Fit')
plt.title(f"Exponential Curve Fitting of y = {a:.3f} $e^{{{b:.3f}x}}$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
