```python
def linear_fit(x, y):
    n = len(x)
    
    sumx = sum(x)
    sumy = sum(y)
    sumx2 = sum(xi * xi for xi in x)
    sumxy = sum(x[i] * y[i] for i in range(n))

    d = (n * sumx2) - (sumx * sumx)
    d1 = (sumy * sumx2) - (sumx * sumxy)
    d2 = (n * sumxy) - (sumx * sumy)

    if d == 0:
        print("Error: Division by zero encountered in calculation.")
        return None

    a = d1 / d
    b = d2 / d

    print(f"The best fit line is: y = {a:.2f} + {b:.2f}x")
    return a, b

# Input data
x = [50, 70, 100, 120]
y = [12, 15, 21, 25]

# Call the function
linear_fit(x, y)
```

```python
import matplotlib.pyplot as plt
a, b = linear_fit(x, y)

# Generate fitted line values
x_fit = [min(x), max(x)]
y_fit = [a + b * xi for xi in x_fit]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label="Data Points")
plt.plot(x_fit, y_fit, color='red', label="linear Fit Line")
plt.title("Linear Curve Fitting")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
