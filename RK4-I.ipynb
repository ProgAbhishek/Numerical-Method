```python
def f(x,y):
    return x-y

def rk4(xo,yo,h,xn):
    x_val = [xo]
    y_val = [yo]

    while xo<xn:
        k1=h*f(xo,yo)
        k2=h*f(xo+h/2,yo+k1/2)
        k3=h*f(xo+h/2,yo+k2/2)
        k4=h*f(xo+h,yo+k3)

        yo=yo+(k1+2*k2+2*k3+k4)/6
        xo=xo+h

        x_val.append(xo)
        y_val.append(yo)
    return x_val, y_val

xo=float(input("Enter the initial x (xo) : "))
yo=float(input("Enter the initial y (yo) : "))
h=float(input("Enter the step size (h) : "))
xn=float(input("Enter the final x (xn) : "))

x_val, y_val =rk4(xo,yo,h,xn)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_val,y_val,label="RK4 Approximation",color="blue")
#plt.axhline(0,color='gray',linestyle='--')
plt.title('soln of dy/dx = x+y using RK4 method')
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```
