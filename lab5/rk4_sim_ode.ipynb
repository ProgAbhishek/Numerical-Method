```python
import matplotlib.pyplot as plt

# Define the system of ODEs
def f1(x, y, z):
    return z

def f2(x, y, z):
    return -y

def rk4_system(x0, y0, z0, h, xn):
    x_vals = [x0]
    y_vals = [y0]
    z_vals = [z0]

    while x0 < xn:
        k1 = h * f1(x0, y0, z0)
        l1 = h * f2(x0, y0, z0)

        k2 = h * f1(x0 + h/2, y0 + k1/2, z0 + l1/2)
        l2 = h * f2(x0 + h/2, y0 + k1/2, z0 + l1/2)

        k3 = h * f1(x0 + h/2, y0 + k2/2, z0 + l2/2)
        l3 = h * f2(x0 + h/2, y0 + k2/2, z0 + l2/2)

        k4 = h * f1(x0 + h, y0 + k3, z0 + l3)
        l4 = h * f2(x0 + h, y0 + k3, z0 + l3)

        y0 = y0 + (k1 + 2*k2 + 2*k3 + k4) / 6
        z0 = z0 + (l1 + 2*l2 + 2*l3 + l4) / 6
        x0 = x0 + h

        x_vals.append(x0)
        y_vals.append(y0)
        z_vals.append(z0)

    return x_vals, y_vals, z_vals

# Initial conditions
x0 = 0
y0 = 0
z0 = 1  # initial dy/dx
h = 0.1
xn = 10

# Solve using RK4
x_vals, y_vals, z_vals = rk4_system(x0, y0, z0, h, xn)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label="y(x)")
plt.plot(x_vals, z_vals, label="z(x) = dy/dx")
plt.title("RK4 for Simultaneous ODEs")
plt.xlabel("x")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```
